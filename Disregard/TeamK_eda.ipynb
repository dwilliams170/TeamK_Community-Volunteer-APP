```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nonprofits_df = pd.read_csv('../_Community Connection App_ Mock Data - fake_nonprofit_volunteer_opportunities.csv')
nonprofits_df.head()
```

```python
# Renaming columns for clarity
nonprofits_df.columns = ['nonprofit name', 'nonprofit opportunity', 'nonprofit available days', 'nonprofit time of day', 'nonprofit location type', 'nonprofit zipcode']
nonprofits_df.head()
```

```python
# change 'nonprofit zipcode' to int type - zero would represent the volunteer opportunity that are remote
nonprofits_df['nonprofit zipcode'] = nonprofits_df['nonprofit zipcode'].fillna(0).astype(int)
```

```python
# combining the nonprofits with the zipcodes into a list
nonprofit_zipcode = nonprofits_df.groupby('nonprofit name')['nonprofit zipcode'].apply(list).reset_index()
print(nonprofit_zipcode)
```
